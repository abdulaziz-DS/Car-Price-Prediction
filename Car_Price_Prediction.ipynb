```python
import gradio as gr
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/BIA/Dataset/CarPrice_Assignment.csv')

# Drop unnecessary columns
X = df.drop(['car_ID', 'CarName', 'price'], axis=1)
y = df['price']

# Label Encoding
le = LabelEncoder()
X['fueltype'] = le.fit_transform(X['fueltype'])
X['aspiration'] = le.fit_transform(X['aspiration'])
X['carbody'] = le.fit_transform(X['carbody'])
X['drivewheel'] = le.fit_transform(X['drivewheel'])
X['enginelocation'] = le.fit_transform(X['enginelocation'])
X['enginetype'] = le.fit_transform(X['enginetype'])
X['cylindernumber'] = le.fit_transform(X['cylindernumber'])
X['fuelsystem'] = le.fit_transform(X['fuelsystem'])
X['doornumber'] = X['doornumber'].replace({'two': 2, 'four': 4})

# Train the Random Forest model
rf = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rf.fit(X, y)

# Define the prediction function
def predict_car_price(
    fueltype, aspiration, doornumber, carbody, drivewheel, enginelocation,
    wheelbase, carlength, carwidth, carheight, curbweight, enginetype,
    cylindernumber, enginesize, fuelsystem, boreratio, stroke, compressionratio,
    horsepower, peakrpm, citympg, highwaympg
):
    # Create a DataFrame from the input values
    input_data = pd.DataFrame({
        'fueltype': [fueltype],
        'aspiration': [aspiration],
        'doornumber': [doornumber],
        'carbody': [carbody],
        'drivewheel': [drivewheel],
        'enginelocation': [enginelocation],
        'wheelbase': [wheelbase],
        'carlength': [carlength],
        'carwidth': [carwidth],
        'carheight': [carheight],
        'curbweight': [curbweight],
        'enginetype': [enginetype],
        'cylindernumber': [cylindernumber],
        'enginesize': [enginesize],
        'fuelsystem': [fuelsystem],
        'boreratio': [boreratio],
        'stroke': [stroke],
        'compressionratio': [compressionratio],
        'horsepower': [horsepower],
        'peakrpm': [peakrpm],
        'citympg': [citympg],
        'highwaympg': [highwaympg]
    })

    # Predict using the trained model
    prediction = rf.predict(input_data)
    return f"Predicted Car Price: ${prediction[0]:.2f}"

# Define the Gradio interface
iface = gr.Interface(
    fn=predict_car_price,
    inputs=[
        gr.Dropdown(choices=[("Gas", 1), ("Diesel", 0)], label="Fuel Type"),
        gr.Dropdown(choices=[("Standard", 1), ("Turbo", 0)], label="Aspiration"),
        gr.Dropdown(choices=[("Two", 2), ("Four", 4)], label="Number of Doors"),
        gr.Dropdown(choices=[("Convertible", 0), ("Hardtop", 1), ("Hatchback", 2), ("Sedan", 3), ("Wagon", 4)], label="Car Body"),
        gr.Dropdown(choices=[("4WD", 0), ("FWD", 1), ("RWD", 2)], label="Drive Wheel"),
        gr.Dropdown(choices=[("Front", 1), ("Rear", 0)], label="Engine Location"),
        gr.Slider(minimum=86.6, maximum=120.9, label="Wheelbase"),
        gr.Slider(minimum=141.1, maximum=208.1, label="Car Length"),
        gr.Slider(minimum=60.3, maximum=72.3, label="Car Width"),
        gr.Slider(minimum=47.8, maximum=59.8, label="Car Height"),
        gr.Slider(minimum=1488, maximum=4066, label="Curb Weight"),
        gr.Dropdown(choices=[("DOHC", 0), ("DOHCV", 1), ("L", 2), ("OHC", 3), ("OHCF", 4), ("OHCV", 5), ("ROTOR", 6)], label="Engine Type"),
        gr.Dropdown(choices=[("Eight", 0), ("Five", 1), ("Four", 2), ("Six", 3), ("Three", 4), ("Twelve", 5), ("Two", 6)], label="Number of Cylinders"),
        gr.Slider(minimum=61, maximum=326, label="Engine Size"),
        gr.Dropdown(choices=[("1bbl", 0), ("2bbl", 1), ("4bbl", 2), ("idi", 3), ("mfi", 4), ("mpfi", 5), ("spdi", 6), ("spfi", 7)], label="Fuel System"),
        gr.Slider(minimum=2.54, maximum=3.94, label="Bore Ratio"),
        gr.Slider(minimum=2.07, maximum=4.17, label="Stroke"),
        gr.Slider(minimum=7.0, maximum=23.0, label="Compression Ratio"),
        gr.Slider(minimum=48, maximum=288, label="Horsepower"),
        gr.Slider(minimum=4150, maximum=6600, label="Peak RPM"),
        gr.Slider(minimum=13, maximum=49, label="City MPG"),
        gr.Slider(minimum=16, maximum=54, label="Highway MPG")
    ],
    outputs="text",
    title="Car Price Prediction",
    description="Enter the car's features to predict its price."
)

# Launch the interface
iface.launch()
```

```python
import gradio as gr
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
import joblib

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/BIA/Dataset/CarPrice_Assignment.csv')

# Drop unnecessary columns
X = df.drop(['car_ID', 'CarName', 'price'], axis=1)
y = df['price']

# Preprocess categorical columns
categorical_cols = ['fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginelocation',
                    'enginetype', 'cylindernumber', 'fuelsystem']
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le

# Handle 'doornumber' separately
X['doornumber'] = X['doornumber'].replace({'two': 2, 'four': 4})

# Train the Random Forest model
rf = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rf.fit(X, y)

# Save the model to disk
joblib.dump(rf, 'random_forest_model.joblib')

# Load the model (use this in production instead of retraining)
rf = joblib.load('random_forest_model.joblib')

# Define the prediction function
def predict_car_price(
    fueltype, aspiration, doornumber, carbody, drivewheel, enginelocation,
    wheelbase, carlength, carwidth, carheight, curbweight, enginetype,
    cylindernumber, enginesize, fuelsystem, boreratio, stroke, compressionratio,
    horsepower, peakrpm, citympg, highwaympg
):
    # Map categorical inputs using precomputed encodings
    input_data = np.array([
        label_encoders['fueltype'].transform([fueltype])[0],
        label_encoders['aspiration'].transform([aspiration])[0],
        doornumber,
        label_encoders['carbody'].transform([carbody])[0],
        label_encoders['drivewheel'].transform([drivewheel])[0],
        label_encoders['enginelocation'].transform([enginelocation])[0],
        wheelbase, carlength, carwidth, carheight, curbweight,
        label_encoders['enginetype'].transform([enginetype])[0],
        label_encoders['cylindernumber'].transform([cylindernumber])[0],
        enginesize,
        label_encoders['fuelsystem'].transform([fuelsystem])[0],
        boreratio, stroke, compressionratio, horsepower, peakrpm, citympg, highwaympg
    ]).reshape(1, -1)

    # Predict using the trained model
    prediction = rf.predict(input_data)
    return f"Predicted Car Price: ${prediction[0]:.2f}"

# Dynamically generate dropdown choices
fueltype_choices = [(val, idx) for idx, val in enumerate(df['fueltype'].unique())]
aspiration_choices = [(val, idx) for idx, val in enumerate(df['aspiration'].unique())]
doornumber_choices = [("Two", 2), ("Four", 4)]
carbody_choices = [(val, idx) for idx, val in enumerate(df['carbody'].unique())]
drivewheel_choices = [(val, idx) for idx, val in enumerate(df['drivewheel'].unique())]
enginelocation_choices = [(val, idx) for idx, val in enumerate(df['enginelocation'].unique())]
enginetype_choices = [(val, idx) for idx, val in enumerate(df['enginetype'].unique())]
cylindernumber_choices = [(val, idx) for idx, val in enumerate(df['cylindernumber'].unique())]
fuelsystem_choices = [(val, idx) for idx, val in enumerate(df['fuelsystem'].unique())]

# Define the Gradio interface
iface = gr.Interface(
    fn=predict_car_price,
    inputs=[
        gr.Dropdown(choices=fueltype_choices, label="Fuel Type"),
        gr.Dropdown(choices=aspiration_choices, label="Aspiration"),
        gr.Dropdown(choices=doornumber_choices, label="Number of Doors"),
        gr.Dropdown(choices=carbody_choices, label="Car Body"),
        gr.Dropdown(choices=drivewheel_choices, label="Drive Wheel"),
        gr.Dropdown(choices=enginelocation_choices, label="Engine Location"),
        gr.Slider(minimum=df['wheelbase'].min(), maximum=df['wheelbase'].max(), label="Wheelbase"),
        gr.Slider(minimum=df['carlength'].min(), maximum=df['carlength'].max(), label="Car Length"),
        gr.Slider(minimum=df['carwidth'].min(), maximum=df['carwidth'].max(), label="Car Width"),
        gr.Slider(minimum=df['carheight'].min(), maximum=df['carheight'].max(), label="Car Height"),
        gr.Slider(minimum=df['curbweight'].min(), maximum=df['curbweight'].max(), label="Curb Weight"),
        gr.Dropdown(choices=enginetype_choices, label="Engine Type"),
        gr.Dropdown(choices=cylindernumber_choices, label="Number of Cylinders"),
        gr.Slider(minimum=df['enginesize'].min(), maximum=df['enginesize'].max(), label="Engine Size"),
        gr.Dropdown(choices=fuelsystem_choices, label="Fuel System"),
        gr.Slider(minimum=df['boreratio'].min(), maximum=df['boreratio'].max(), label="Bore Ratio"),
        gr.Slider(minimum=df['stroke'].min(), maximum=df['stroke'].max(), label="Stroke"),
        gr.Slider(minimum=df['compressionratio'].min(), maximum=df['compressionratio'].max(), label="Compression Ratio"),
        gr.Slider(minimum=df['horsepower'].min(), maximum=df['horsepower'].max(), label="Horsepower"),
        gr.Slider(minimum=df['peakrpm'].min(), maximum=df['peakrpm'].max(), label="Peak RPM"),
        gr.Slider(minimum=df['citympg'].min(), maximum=df['citympg'].max(), label="City MPG"),
        gr.Slider(minimum=df['highwaympg'].min(), maximum=df['highwaympg'].max(), label="Highway MPG")
    ],
    outputs="text",
    title="Car Price Prediction",
    description="Enter the car's features to predict its price."
)

# Launch the interface
iface.launch()
```
